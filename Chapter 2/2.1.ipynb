## Difined Unit Catalog to use

```python
%sql
USE CATALOG hive_metastore
```

## Create table

```python
%sql
Create Table if not exists product_info(
    product_id INT,
    product_name STRING,
    categry STRING,
    prince Double,
    quantity INT
)
```

## Inserting Data

```python
%sql
insert into product_info(product_id,product_name,categry,prince,quantity) 
values
(1,'Laptop','Electronics',1000,100);

insert into product_info(product_id,product_name,categry,prince,quantity) 
values
(2,'Mobile','Electronics',500,200),
(3,'Shirt','Clothing',200,135),
(4,'Pants','Clothing',300,150);
```

```python
%sql
select * from product_info;
```
