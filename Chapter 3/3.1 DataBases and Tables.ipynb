## Create managed Table

```python
%sql
create table if not exists default.managed_default(
  country string,
  code string,
  dial_code String
);

insert into managed_default values('India','IN','91');
select * from managed_default
```

```python
%sql
describe extended managed_default
```

## Create a External Table

```python
%sql
drop table external_default;
CREATE TABLE IF NOT EXISTS external_default (
  country string,
  code string,
  dial_code String
)
location 'dbfs:/mnt/demo/external_default';

insert into external_default 
values('India','IN','91');

select * from external_default
```

```python
%sql
describe extended external_default;
```
