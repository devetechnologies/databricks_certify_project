```python
%sql
create table if not exists  students 
 as select * from json.`${dataset.school}/students-json`;

 describe extended students;
    
```

```python
%sql
create table courses_unparsed as select * from csv.`${dataset.school}/courses-csv`;
 describe extended courses_unparsed;
```

```python
%sql
select * from students;
```

```python
%sql
select * from courses_unparsed;
```

```python
%sql
create table courses_csv(
  course_id String,
  title String,
  instructor String,
  category String,
  price Double
)
using csv
options (
  header = true,
  delimiter = ';'
)
location "${dataset.school}/courses-csv"
```

```python
%sql
select * from courses_csv
```

```python
%sql
describe extended courses_csv
```

```python
%sql
select * from csv.`${dataset.school}/courses-csv/export_*.csv`;
```

```python
%sql
create or replace temp view courses_tmp_vw(
  course_id String,
  title String,
  instructor String,
  category String,
  price Double
)
USING CSV
options (
  path = '${dataset.school}/courses-csv/export_*.csv',
  header = true,
  delimiter = ';'
);

create table if not exists courses
as select * from courses_tmp_vw
```

```python
%sql
select * from courses;
```
