## Create UDF Function


```python
%sql
create or replace function get_letter_grade(gpa double) 
returns string 
return case
  when gpa >= 3.5 then 'A'
  when gpa >= 2.75 and gpa < 3.5 then 'B' 
  when gpa >= 2.0 and gpa < 2.75 then 'C'
  else 'F'
end;
```

```python
%sql
DESCRIBE FUNCTION EXTENDED get_letter_grade
```
